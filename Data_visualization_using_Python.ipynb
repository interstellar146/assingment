```python
### Data visualization
### library Matplotib
import matplotlib.pyplot as plt
```

```python
### Line Plot
x1 = [1,2,3,4,5]
y1 = [2,5,2,6,8]
x2 = [1,2,3,4,5]
y2 = [4,5,8,9,10]
plt.xlabel("X Axis",fontsize=12,fontstyle='italic')
plt.ylabel("Y Axis",fontsize=12)
plt.title("Line Plot",fontsize=15,fontname='Courier')
plt.plot(x1,y1,color="red",label="First Graph")   ### line plot
plt.plot(x2,y2,color="blue",label="Second Graph")   ### line plot
plt.legend(loc=2)
plt.grid()
#plt.axis('off')
plt.show()
```

```python
# Bar plot
import matplotlib.pyplot as plt

# Look at index 4 and 6, which demonstrate overlapping cases.
x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 2, 4, 7, 8, 3]
x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]

plt.bar(x1, y1, label="Blue Bar", color='b')
plt.bar(x2, y2, label="Green Bar", color='g')
plt.plot()

plt.xlabel("bar number")
plt.ylabel("bar height")
plt.title("Bar Graph")
plt.legend()
plt.show()
```

```python
#### Scatter Plot
x1 = [1,2,3,4,5]
y1 = [2,5,2,6,8]
x2 = [1,2,3,4,5]
y2 = [4,5,8,9,10]
plt.xlabel("X Axis",fontsize=12,fontstyle='italic')
plt.ylabel("Y Axis",fontsize=12)
plt.title("Scatter Plot",fontsize=15,fontname='Courier')
plt.scatter(x1,y1,color="red",label="First Graph")   ### line plot
plt.scatter(x2,y2,color="blue",s=50,marker="*",label="Second Graph")   ### line plot
plt.plot(x2,y2,color="blue")
plt.legend(loc=2)
#plt.axis('off')
plt.show()
```

```python
### Histogtram
import numpy as np
sample = np.random.randint(10,100,30)
plt.title("Histogtram  Plot",fontsize=15,fontname='Courier')
plt.hist(sample,rwidth=0.7)
plt.show()
```

```python
### Pie Chart
plt.figure(figsize=(7,7))
slices = [10,20,50,30,34]
act = ["S1","S2","S3","S4","S5"]
cols = ["red","blue","green","pink","yellow"]
plt.title("Pie Chart",fontsize=15,fontname='Courier')
plt.pie(slices,labels=act,colors=cols,
        autopct="%1.3f%%",explode=(0,0.6,0,0.4,0))
plt.show()
```
